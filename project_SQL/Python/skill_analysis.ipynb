```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('C:/Users/dalys/OneDrive/Desktop/SQL_Data_Analysis/project_SQL/top_data_analyst_skills.csv')
skill_counts = df['skills'].value_counts()
```

```python
df.head()
```

```python
#VISUALIZE THE TOP SKILLS USING SEABORN
sns.set_style('dark')

plt.figure(figsize=(10,6), facecolor= 'black')
ax = sns.barplot(x=skill_counts.values[:10], y=skill_counts.index[:10], palette = 'coolwarm')
ax.set_facecolor('black')
#Labels and title

ax.set_xlabel('Number of Jobs', color = 'white')   
ax.set_ylabel('Skills', color = 'white')
ax.set_title('Top 10 Skills for Data Analyst Jobs', color = 'white')
ax.tick_params(colors='white')
plt.show()

```

# Data Analyst Skills Analysis
As we can see, the top 10 skills for data analyst jobs are SQL, Python, R, Express, Excel, AWS, Pandas, ElasticSearch, Neo4j, and SQL Server.
Python and SQL are the most popular skills for data analyst jobs as they both appears in 6 out of the top 10 job postings. One surprising skill that was mentioned in these job
postings is Express. Express is a web application framework for Node.js, which is a runtime environment that executes JavaScript code outside a web browser. It might be worhthwhile to 
look deeper into the data to find out why Express is a popular skill for data analyst jobs or if it is a mistake in the job postings.
