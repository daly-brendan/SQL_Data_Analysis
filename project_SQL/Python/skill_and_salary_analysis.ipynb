```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
da_skills = pd.read_csv('/Users/dalys/OneDrive/Desktop/SQL_Data_Analysis/project_SQL/CSVs/highest_paying_skills_DA.csv')
ds_skills = pd.read_csv('/Users/dalys/OneDrive/Desktop/SQL_Data_Analysis/project_SQL/CSVs/highest_paying_skills_DS.csv')
```

```python
#VISUALIZE THE HIGHEST PAYING SKILLS FOR DATA ANALYSTS
sns.set_style('dark')

plt.figure(figsize=(10,6), facecolor= 'black')
ax = sns.barplot(x=da_skills.average_salary, y=da_skills.skills, palette = 'coolwarm')
ax.set_facecolor('black')
#Labels and title

ax.set_xlabel('Average Salary per Skill($)', color = 'white')   
ax.set_ylabel('Skills', color = 'white')
ax.set_title('Highest Paying Skills for Data Analysts', color = 'white')
ax.tick_params(colors='white')
plt.show()
```

# Highest Paying Data Analyst Skills

As you can see right away from this graph there is a large outlier - the version control system SVN has an average salary of $400,000. If you compare the average between the data analyst skills and the data science skills they are roughy the same with DS leading by 20,000 annually (145,000 vs 167,000). 

I was shocked to see a lot of AI/machine learning tools mentioned in this list like HuggingFace, tensorflow, pytorch, and keras as that is something I would expect a data scientist be required to know and not a data analyst.

Its also interesting that the most common data analysis tools such as SQL, python, or excel are not mentioned in this top 25 which might mean that the highest paying data analyst roles are more specialized than a normal data or business analyst.

```python
# MOST OPTIMAL SKILLS ANALYSIS

df3 = pd.read_csv('C:/Users/dalys/OneDrive/Desktop/SQL_Data_Analysis/project_SQL/CSVs/most_optimal_skills.csv')
```

```python
sns.set_style('dark')

fig, ax = plt.subplots(figsize=(10,6), facecolor= 'black')
ax.set_facecolor('black')

scatter = ax.scatter (
    df3['average_salary'], df3['skills'],
    s = df3['demand_count']*10,
    alpha = 0.7,
    color = 'red',
    edgecolors='white'
)

ax.set_xlabel('Average Salary per Skill($)', color = 'white')
ax.set_ylabel('Skills', color = 'white')
ax.set_title('Most Optimal Skills for Data Analysts and Data Scientists', color = 'white')

ax.tick_params(colors='white')
ax.grid(color = 'gray', linestyle = '--', linewidth = 0.5)

plt.show()
```

# Most Optimal Skills for Data Analysts and Data Scientists

The bubble chart visualizes the most optimal skills for data analysts and data scientists based on average salary and demand count. Each bubble represents a specific skill, with the size of the bubble corresponding to the demand count for that skill. 

- **X-axis**: Average Salary per Skill ($)
- **Y-axis**: Skills
- **Bubble Size**: Demand Count (larger bubbles indicate higher demand)

From the chart, we can observe the following:

1. **High Demand and High Salary**: Skills like `snowflake`, `spark`, and `databricks` have large bubbles indicating high demand and also offer competitive salaries.
2. **Specialized Skills**: Some skills like `kafka` and `pytorch` have smaller bubbles but still offer high salaries, indicating they are specialized skills with moderate demand.
3. **General Trends**: The chart helps identify which skills are both in high demand and offer high salaries, making them optimal for professionals looking to maximize their earning potential.

Overall, this visualization provides valuable insights into the most lucrative and sought-after skills in the data analysis and data science fields.
